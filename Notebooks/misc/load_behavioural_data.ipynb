# How to load behavioural data from for your own analyses

GridMaze data is organised in a stardarised format, see (notebook_that_will_soon_exist.ipynb) for details on how the data is organised on disk & how we access it for the analyses in this repo.

In short, we organise all data at the subject/session level and access the data we want using a ```get_maze_sessions``` function that lives in ```GridMaze.analysis.core.get_sessions```. However, if you'd like to load all the behavioural data in a single table/dataframe for your own analysis, use the code below :)

You can load behavioural data at either node-by-node resolution or frame-by-frame resolution. 

Note:
- Run the code below in a conda environment with the packages specified in ```GridMaze_mFC_refactor/code/requirements.txt```
- Recommended have >32GB RAM for loading frame-by-frame data.

### Load node-by-node data

````python
import pandas as pd
from GridMaze.analysis.core import get_sessions as gs

# load data
data_structures = ["trajectory_decisions_df"]  # load trajectory_decisions_df for node-by-node behaavioural data

sessions = gs.get_maze_sessions(
    subject_IDs="all",
    maze_names="all",
    days_on_maze="all",
    with_data=data_structures,
    must_have_data=True,
)

# combine data across sessions into a single dataframe
behaviour_df = pd.concat([session.trajectory_decisions_df for session in sessions])

behaviour_df.head()
````

````python
# save where you want
save_path = "your/path/here"
behaviour_df.to_csv(save_path, index=False)
````

### Load frame-by-frame data

Note we can load this with or without route (Xiao™) info

````python
import pandas as pd
from GridMaze.analysis.core import get_sessions as gs

# load data with specified inputs
with_routes = True
data_structures = ["navigation_df", "navigation_routes_df"] if with_routes else ["navigation_df"]

sessions = gs.get_maze_sessions(  # loading alot of data into memory, will take a few mins (recommend 128GB RAM)
    subject_IDs="all",
    maze_names="all",
    days_on_maze="all",
    with_data=data_structures,
    must_have_data=True,
)
````

````python
# combine data across sessions into a single dataframe
if with_routes:
    behaviour_dfs = []
    for s in sessions:
        behaviour_dfs.append(pd.concat([s.navigation_df, s.navigation_routes_df.reset_index(drop=True)], axis=1))
else:
    behaviour_dfs = [s.navigation_df for s in sessions]

behaviour_df = pd.concat(behaviour_dfs, axis=0, ignore_index=True)

behaviour_df.head()
````

````python
# save where you want
save_path = "your/path/here"
behaviour_df.to_csv(save_path, index=False)
````
